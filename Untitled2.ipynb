#ðŸ”¸ Average Price by Pizza Category

```python
df.columns
```

```python
query="""SELECT categories ,city,"menus.amountMax"
FROM sales
ORDER BY "menus.amountMax" DESC"""
pd.read_sql(query, conn)
```

```python
query="""SELECT categories, ROUND(AVG("menus.amountMax"),2) AS AVG_PRICE FROM sales GROUP BY categories
ORDER BY "AVG PRICE" DESC"""
pd.read_sql(query, conn)
```
